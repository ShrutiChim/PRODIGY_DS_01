```python
import pandas as pd
import matplotlib.pyplot as plt
df_main = pd.read_csv('C:/Users/shrut/OneDrive/Desktop/New folder/API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv', delimiter=',', skiprows=4)
print(df_main.columns)
df_population_2021 = df_main[['Country Name', '2021']]
df_population_2021 = df_population_2021.dropna(subset=['2021'])
df_population_2021['2021'] = pd.to_numeric(df_population_2021['2021'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.hist(df_population_2021['2021'], bins=20, edgecolor='black')
plt.title('Distribution of Population in 2021')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
